```python
import torch
import matplotlib.pyplot as plt
```

```python
# just tensor
A = torch.arange(-10, 10, 1, dtype=torch.float32)
A
```

```python
plt.plot(A);
```

```python
# Create ReLU function by hand
def relu(x):
  return torch.maximum(torch.tensor(0), x) # inputs must be tensors

# Pass toy tensor through ReLU function
relu(A)
```

```python
plt.plot(relu(A));
```

```python
# Create a custom sigmoid function
def sigmoid(x):
  return 1 / (1 + torch.exp(-x))

# Test custom sigmoid on toy tensor
sigmoid(A)
```

```python
plt.plot(sigmoid(A));
```

```python
def leaky_relu(x, alpha=2):
    return torch.maximum(alpha * x, x)

leaky_relu(A)
```

```python
plt.plot(leaky_relu(A))
```

```python
def elu(x, alpha=1.0):
    return torch.where(x >= 0, x, alpha * (torch.exp(x) - 1))

elu(A)
```

```python
plt.plot(elu(A))
```

```python
def tanh(x):
    return torch.tanh(x)

tanh(A)
```

```python
plt.plot(tanh(A))
```

```python
def selu(x, alpha=1.67326, scale=1.0507):
    return scale * torch.where(x >= 0, x, alpha * (torch.exp(x) - 1))

selu(A)
```

```python
plt.plot(selu(A))
```

```python
def softmax(x):
    return torch.exp(x) / torch.sum(torch.exp(x), dim=0, keepdim=True)

softmax(A.reshape(-1, 1))
```

```python
plt.plot(softmax(A.reshape(-1, 1)))
```
